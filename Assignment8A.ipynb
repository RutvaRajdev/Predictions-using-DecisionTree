# The Dataset

I have chosen the <b> Teaching Assistant Evaluation dataset </b> available on the UCI Machine Learning repository. It can be found at http://archive.ics.uci.edu/ml/datasets/Teaching+Assistant+Evaluation.
This dataset consists of evaluations of 151 teaching assistants over three regular semesters and two summer semesters at University of Wisconsin-Madison. <br> <br>
The dataset consists of 151 instances and 5 attributes. The attributes are both numerical and categorical. There is no missing data.<br>
#### Attribute Information
The data consists of 5 attributes as follows: <br>
Binary: Whether or not the TA is a native English speaker (1 for English speaker, 2 for non english speaker) <br>
Categorical: Course Instructor (25 categories) <br>
Categorical: Course (26 categories) <br>
Binary: Summer or regular semester (1 for summer, 2 for regular) <br>
Numerical: Class size <br>

The class attribute 'Score' denotes the performance score. It is a categorical variable. 1 denotes low score, 2 denotes medium score while 3 denotes high score.

# Dataset Analysis

The dataset does not have any missing data. It has 5 features, 4 of which are categorical features while 1 is numerical. Out of the 4 categorical features, 2 are binary. 

```python
import pandas as pd
# Loading dataset
fileURL = "http://archive.ics.uci.edu/ml/machine-learning-databases/tae/tae.data"
df = pd.read_csv(fileURL, names=["English Speaker", "Course Instructor", "Course", "Semester", "Class Size", "Score"], header=None)
feature_names = ["English Speaker", "Course Instructor", "Course", "Semester", "Class Size"]
class_names = ["Low", "Medium", "High"]

# A sneak peek into the dataset
df.head()
```

# Implementing Decision Tree

```python
# Differentiating data and target variables from dataset
data = df[["English Speaker", "Course Instructor", "Course", "Semester", "Class Size"]]
target = df[["Score"]]

# Training a supervised learning model
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(data, target)

# Generating the tree visualization
import graphviz 

dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=feature_names,  
                         class_names=class_names,  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
# Creating a new instance
# Taking the Teaching Assistant feature as 1 (English Speaker)
# Taking Course Instructor category as 15
# Taking Course category as 25
# Taking semester type as 2 (Regular semester)
# Taking class size as 30

clf.predict([[1, 15, 25, 2, 30]])
```

Our trained model predicted the performance score as 3 i.e. high which actually makes sense and we can trace the result back to our tree model also. <br>
The tree starts with the root node where it checks whether the course value is less than or equal to 5.5. In our instance it is not. Thus it would go to the right branch. <br>
Here it would check if the course instructor value is less than equal to 19. It is 15 in our instance and thus it would proceed towards the left branch. <br>
Next we would check if the course value is less than or equal to 13. This condition is false in our case since the course value is 25 and thus we would proceed on the right branch. <br>
Here we would directly get the answer as high which is value 3. <br>
We got the same answer when we used our model for prediction.
